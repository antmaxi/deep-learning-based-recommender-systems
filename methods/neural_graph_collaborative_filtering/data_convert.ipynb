```python
import pandas as pd

from random import seed
from random import randint

def output_dict(data_dict, out_name):
    with open(out_name, 'w+') as out_file:
        for n in data_dict.keys():
            if not data_dict[n] is None:
                out_file.write(str(n))
                out_file.write(' ')
                listToStr = ' '.join([str(elem) for elem in data_dict[n]]) 
                out_file.write(listToStr)
                out_file.write('\n')
# seed random number generator
seed(1)

df = pd.read_csv("./Data/ml-1m/1/ratings.dat", 
                 sep="::", 
                 #skiprows=1, 
                 #usecols=[0,7], 
                names=['userID','movieID', 'rating', 'timestamp']
                )

users = set(range(1,100))#df['userID'])
usersID = list(users)
moviesID_train = []
moviesID_test = []

max_movieID = max(df['movieID'])
k = 100
for n in users:
    l = list(df[df['userID'] == n]['movieID'])
    m = max(l)
    current_test = [m]
    while len(current_test) < k:
        value = randint(0, max_movieID)
        if (value not in current_test) and (value not in l):
            current_test.append(value)
    moviesID_test.append(current_test)
    l.remove(m) 
    moviesID_train.append(l)
    
dict_test = dict(zip(usersID, moviesID_test))
dict_train = dict(zip(usersID, moviesID_train))
print(dict_train)
out_test = './Data/ml-1m_light/test.txt'
out_train = './Data/ml-1m_light/train.txt'
output_dict(dict_test, out_test)
output_dict(dict_train, out_train)
```

# Reformatting Nicolas'es data

```python
import pandas as pd

from random import seed
from random import randint

name = 'jester'

def output_dict(data_dict, out_name):
    with open(out_name, 'w+') as out_file:
        for n in data_dict.keys():
            if not data_dict[n] is None:
                out_file.write(str(n))
                out_file.write(' ')
                listToStr = ' '.join([str(elem) for elem in data_dict[n]]) 
                out_file.write(listToStr)
                out_file.write('\n')
# seed random number generator
seed(1)

path = "./Data/" + name + "/"
df_test = pd.read_csv(path + name + ".test.rating", 
                 sep="\t",  
                names=['userID','movieID', 'rating', 'timestamp']
                )
df_train = pd.read_csv(path + name + ".train.rating", 
                 sep="\t",  
                names=['userID','movieID', 'rating', 'timestamp']
                )
df_neg = pd.read_csv(path + name + ".test.negative", 
                 sep="\t",
                 header=None
                )

users = set(df_train['userID'])
usersID = list(users)
moviesID_train = []
moviesID_test = []

for n in users:
    
    current_test = list(df_test[df_test['userID'] == n]['movieID'])
    l_neg = list(df_neg.loc[n, df_neg.columns != 0])
    current_test += l_neg
    #print(n, current_test)
    moviesID_test.append(current_test)
    
    l = list(df_train[df_train['userID'] == n]['movieID'])
    moviesID_train.append(l)
    
dict_test = dict(zip(usersID, moviesID_test))
dict_train = dict(zip(usersID, moviesID_train))
print(dict_train)
out_test = path + 'test.txt'
out_train = path + 'train.txt'
output_dict(dict_test, out_test)
output_dict(dict_train, out_train)
```
